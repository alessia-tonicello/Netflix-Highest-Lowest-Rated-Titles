PROJECT: analyzing Netflix's Top 10 Highest and Lowest Rated Titles

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#We load the dataset
titles = pd.read_csv('Netflix_Engagement.csv')

# Look at the first few rows of the dataset
print(titles.head())
```

Now there is an important step that is Data Cleaning, because if for exmaple a movie’s rating is missing, we may decide to remove that row, or we might fill the missing value with the average or median ratin.

```python
# Check for missing values
print(titles.isnull().sum())  # Check for missing values in each column

# After having checked for missing values, we can remove the rows with missing values in the 'Rating' and 'Title' columns
df_cleaned = titles.dropna(subset=['Rating', 'Title'])

#We can also check if there are some duplicate rows in the dataset and so we can remove them
df_cleaned = titles.drop_duplicates()
```

After this we can sort the data by rating, in order to identify the top-rated movies. In the context of this project, we want to find the highest-rated movies, so sorting the dataset by the rating column in descending order make sure that the movies with the highest ratings appear at the top of the list.

```python
# Sort the dataset by 'rating' in descending order
df_sorted = titles.sort_values(by='Rating', ascending=False)

# Now we select the top 10 highest-rated movies
top_10_titles = df_sorted.head(10)
print(top_10_titles)
```

After this, as a last step, we can use matplotlib to visualize the data, and in order to do this we'll use a horizontal bar chart 

```python
# Create a horizontal bar chart to visualize the top 10 highest-rated movies, defining all its characteristics
plt.figure(figsize=(10, 6))  
plt.barh(top_10_titles['Title'], top_10_titles['Rating'], color='skyblue') 
plt.xlabel('Rating')  
plt.ylabel('Title')  
plt.title('Top 10 Highest-Rated Titles')  
plt.gca().invert_yaxis() 
plt.show()
```

Thanks to the analysis of this dataset containnig informatino on Netflix titles from January to June 2023, we have been able to see that Big Time Rush seasons are the top reated ones. 

Starting from this, we could also see which are the 10 Least-Rated Titles.

```python
#We start by sorting the list in ascending order
df_sorted = titles.sort_values(by='Rating', ascending=True)

# Now we select the top 10 least-rated movies
top_10_least_rated_titles = df_sorted.head(10)
print(top_10_least_rated_titles)

```

```python
# Now we can visualize our data in the way we did before with the bar chart 
plt.figure(figsize=(10, 6))  
plt.barh(top_10_least_rated_titles['Title'], top_10_least_rated_titles['Rating'], color='skyblue') 
plt.xlabel('Rating')  
plt.ylabel('Title')  
plt.title('Top 10 Least-Rated Titles')  
plt.gca().invert_yaxis() 
plt.show()
```
